```python
import nltk
nltk.download('abc')
nltk.download('brown')
nltk.download('semcor')
nltk.download('words')
```

```python
# Model
# !pip install spello
from spello.model import SpellCorrectionModel

# for data(words)
from nltk.corpus import brown,abc,semcor,words
import re

import pandas as pd
```

```python
abc_words = list(abc.words())
brown_words = list(brown.words())
semcor_words = list(semcor.words())
wo = words.words('en')
words = abc_words + brown_words + semcor_words + wo
```

```python
# removing numbers and single letter(a)
words = [word for word in words if re.match(r'^[a-zA-Z]+$',word) and len(word) > 1]
```

```python
# Model
sp = SpellCorrectionModel(language='en')
```

```python
#train
sp.train(words)
```

```python
# saving the trained model
sp.save(model_save_dir='D:\\trained_model.pkl')
```

```python
df = pd.read_csv('NLP_Chardes_Evaluation.csv')
```

```python
df
```

```python
# predictiong the masked words in evaluation data set
cor = []
for i in df['Masked']:
    i = re.sub(" ","",i)
    cor.append(sp.spell_correct(i)['spell_corrected_text'])
```

```python
df['Actual'] = cor
```
